<h3> Data Cleaning and Re-Formatting </h3>

<h6> Published time and trending date columns are in incorrect format. So, converting them to proper format using datetime </h6>

```python
all_countries_df['trending_date'] = pd.to_datetime(all_countries_df['trending_date'], format='%y.%d.%m')

all_countries_df['publish_time'] = pd.to_datetime(all_countries_df['publish_time'], format='%Y-%m-%dT%H:%M:%S.%fZ')
```

```python
copy_dataframe=all_countries_df.copy()
```

<h6> Cleaning Tags column by splitting, converting to lower case and replacing characters </h6>

```python
def tag_cleaning(tag):
    if tag == '[none]': return []
    return tag.lower().replace('"','').split('|')

all_countries_df['tags'] = all_countries_df['tags'].apply(tag_cleaning)

all_countries_df.head()
```

<h3> Clean Titles </h3>

<h6> In order to clean titles, stopwords library is needed from nltk. Titles are removed for stopwords, numeric and other special characters. A new column called 'title_cl' is appended to the dataframe </h6>

```python
def clean_titles(title):
    tokens = title.lower().split()
    cleaned = []
    stop_word = set(stopwords.words('english'))
    for token in tokens:
        if token.startswith('$') or token.isnumeric() or not token.isalnum() or token in stop_word:  # Remove money amount, non-alphanumeric tokens, or stopwords
            continue
        else:
            cleaned.append(token)
    return ' '.join(cleaned)

all_countries_df['title_cl'] = all_countries_df['title'].apply(clean_titles)
```

<h3> Sentiment analysis for title, description and tags column using WordCloud, TextBlob </h3>

```python
from wordcloud import WordCloud


from nltk import sent_tokenize, word_tokenize
from wordcloud import WordCloud, STOPWORDS
```

```python
from collections import Counter
from nltk.tokenize import RegexpTokenizer
#from stop_words import get_stop_words
```

```python
globaldf=all_countries_df.copy()
top_limit = 100
```

<h3> Cleaning before sentiment analysis </h3>

<h6> STOPWORDS library is used to get the list of stopwords </h6>

```python
a = globaldf['title'].str.lower().str.cat(sep=' ')
b = re.sub('[^A-Za-z]+', ' ', a)
nltk_words = list(stopwords.words('english')) 
word_tokens = word_tokenize(b)

filtered_sentence = [w for w in word_tokens if not w in nltk_words]
filtered_sentence = []
for w in word_tokens:
    if w not in nltk_words and (len(w)>2):
        filtered_sentence.append(w)

cleaned_data_title = [word for word in filtered_sentence if not word.isnumeric()]
```

```python
word_dist = nltk.FreqDist(cleaned_data_title)
rslt = pd.DataFrame(word_dist.most_common(top_limit),
                    columns=['Word', 'Frequency'])

plt.figure(figsize=(10,10))
sns.set_style("whitegrid")
ax = sns.barplot(x="Word",y="Frequency", data=rslt.head(7))
plt.title('Top words in title', fontsize=18)
plt.show()
```

```python
def wc(data,bgcolor,title):
    plt.figure(figsize = (50,50))
    wc = WordCloud(background_color = bgcolor, max_words = 1000,  max_font_size = 50)
    wc.generate(' '.join(data))
    plt.imshow(wc)
    plt.axis('off')
```

<h6> Word Cloud of common words </h6>

```python
wc(cleaned_data_title,'black','Common Words' )
```

<h3> Frequency of words in Tags column </h3>

```python
tags_lower = all_countries_df['tags'].astype(str).str.lower().str.cat(sep=' ')

tags_remove_pun = re.sub('[^A-Za-z]+', ' ', tags_lower)

nltk_words = list(stopwords.words('english')) 

word_tokens_tags = word_tokenize(tags_remove_pun)
filtered_sentence_tags = [w_tags for w_tags in word_tokens_tags if not w_tags in nltk_words]

filtered_sentence_tags = []
for w_tags in word_tokens_tags:
    if w_tags not in nltk_words:
        filtered_sentence_tags.append(w_tags)

words_len=[word for word in filtered_sentence_tags if len(word) > 2]
cleaned_data_tags = [word_tags for word_tags in words_len if not word_tags.isnumeric()] 


word_dist_tags = nltk.FreqDist(cleaned_data_tags)
rslt_tags = pd.DataFrame(word_dist_tags.most_common(top_limit),
                    columns=['Word', 'Frequency'])

plt.figure(figsize=(10,10))
sns.set_style("whitegrid")
ax = sns.barplot(x="Word",y="Frequency", data=rslt_tags.head(7))
plt.title('Top words in tags', fontsize=18)
plt.show()
```

```python
def f(desc_sentiment):
    if desc_sentiment['sentiment'] > 0:
        val = "Positive"
    elif desc_sentiment['sentiment'] == 0:
        val = "Neutral"
    else:
        val = "Negative"
    return val
```

<h6> Cleaning Tags column by splitting, converting to lower case and replacing characters </h6>

```python
def tag_cleaning(tag):
    if tag == '[none]': return []
    return tag.lower().replace('"','').split('|')
```

```python
us_data_dataframe['tags'] = us_data_dataframe['tags'].apply(tag_cleaning)
```

<h3> Clean Titles </h3>

<h6> In order to clean titles, stopwords library is needed from nltk. Titles are removed for stopwords, numeric and other special characters. A new column called 'title_cl' is appended to the dataframe </h6>

```python
nltk.download('stopwords')
```

```python
def clean_titles(title):
    tokens = title.lower().split()
    cleaned = []
    sw = set(stopwords.words('english'))
    for token in tokens:
        if token.startswith('$') or token.isnumeric() or not token.isalnum() or token in sw:  # Remove money amount, non-alphanumeric tokens, or stopwords
            continue
        else:
            cleaned.append(token)
    return ' '.join(cleaned)

us_data_dataframe['title_cl'] = us_data_dataframe['title'].apply(clean_titles)
```

```python
us_data_dataframe.head()
```

<h3> Sentiment analysis for title, description and tags column using WordCloud, TextBlob </h3>

```python
#importing libraries
from wordcloud import WordCloud


from nltk import sent_tokenize, word_tokenize
from wordcloud import WordCloud, STOPWORDS
```

```python
from collections import Counter
from nltk.tokenize import RegexpTokenizer
#from stop_words import get_stop_words
```

```python
usadf=us_data_dataframe.copy()
top_limit = 100
```

```python
nltk.download('punkt')
```

<h3> Cleaning before sentiment analysis </h3>

<h6> STOPWORDS library is used to get the list of stopwords </h6>

```python
#CLeaning Title 

a = usadf['title'].str.lower().str.cat(sep=' ')
b = re.sub('[^A-Za-z]+', ' ', a)
nltk_words = list(stopwords.words('english')) 
word_tokens = word_tokenize(b)

filtered_sentence = [w for w in word_tokens if not w in nltk_words]
filtered_sentence = []
for w in word_tokens:
    if w not in nltk_words:
        filtered_sentence.append(w)

cleaned_data_title = [word for word in filtered_sentence if not word.isnumeric()]
```

```python
word_dist = nltk.FreqDist(cleaned_data_title)
rslt = pd.DataFrame(word_dist.most_common(top_limit),
                    columns=['Word', 'Frequency'])

plt.figure(figsize=(10,10))
sns.set_style("whitegrid")
ax = sns.barplot(x="Word",y="Frequency", data=rslt.head(7))
plt.title('Frequen words in title', fontsize=15)
plt.show()
```

```python
def wc(data,bgcolor,title):
    plt.figure(figsize = (50,50))
    wc = WordCloud(background_color = bgcolor, max_words = 1000,  max_font_size = 50)
    wc.generate(' '.join(data))
    plt.imshow(wc)
    plt.axis('off')
```
